```python
import matplotlib.pyplot as plt
```

```python
labels = ['Springer Link', 'Scopus', 'IEEE Digital Library', 'ACM Digital Library']
slices = [1183, 855, 420, 203]
total = sum(slices)

colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

explode = (.03,.02,.03,.02)

fig1, ax1 = plt.subplots()

plt.pie(slices, labels=labels, colors=colors,
        autopct=lambda p: '{:.0f}'.format(p * total / 100), 
        startangle=90, pctdistance=0.85, explode=explode)

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

ax1.set_facecolor('white')
ax1.axis('equal') 
plt.tight_layout()
plt.show()
```

```python
def circle_plot(data, title='graph'):
    total = sum(data.values())

    colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

#     explode = [ .03 for _ in range(len(slices)) ]

    fig1, ax1 = plt.subplots()

    plt.pie(data.values(), labels=data.keys(), colors=colors, autopct='%1.1f%%', 
            startangle=90, pctdistance=0.85)#, explode=explode[0:len(slices)])

    centre_circle = plt.Circle((0,0),0.70,fc='white')
    fig = plt.gcf()
    fig.gca().add_artist(centre_circle)

    ax1.set_facecolor('white')
    ax1.axis('equal') 
    plt.title(title, fontdict={'fontsize': 'xx-large'})
    plt.tight_layout()
    plt.show()
```

# Artigos aceitos

```python
circle_plot({'Descartados': 443, 'Aceitos': 383}, 'Artigos descartados e aceitos')
```

# Sobre os métodos de avaliação

    Alguns artigos só foram classificados (sim/não), sem coletar as infromações. Outros artigos apresentaram o uso de mais de um método de avaliação, entrando na porcentagem dos dois valores.

```python
circle_plot({'Simulação': 132, 'Aferição': 84, 'Modelagem Analítica': 15}, 'Métodos de avaliação')
```
